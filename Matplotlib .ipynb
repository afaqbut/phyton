## Matplotlib 

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

## prepare

```python
x = np.linspace(0, 4*np.pi, 10)
y = np.sin(x)
print(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
fig.show()
```

## Tweak

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, color='blue')
plt.title('Graph')
plt.xlabel('time')
plt.ylabel('voltage')
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, linestyle="--")
plt.title('Graph')
plt.xlabel('time')
plt.ylabel('voltage')
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x,y, linewidth=5)
plt.title('Graph')
plt.xlabel('time')
plt.ylabel('voltage')
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, marker='o')
plt.title('Graph')
plt.xlabel('time')
plt.ylabel('voltage')
```

## Organize

```python
X = np.linspace(0, 10, 100)
Y1, Y2 = np.sin(x), np.cos(x)
plt.plot(X, Y1, X, Y2)
```

```python
fig, (plt1, plt2) = plt.subplots(2,1)
plt1.plot(X, Y1, color='C1')
plt2.plot(X, Y1, color='C0')
```

```python
fig, (plt1, plt2) = plt.subplots(1,2)
plt1.plot(X, Y1, color='C1')
plt2.plot(X, Y1, color='C0')
```
